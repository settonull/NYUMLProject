```python
import os
import csv

season = '1975'
week = '-1'
week_count=0
teams_seen = []

with open('data/snoozle/snoozle-combined.csv', 'r') as ofile:
    statsr = csv.reader(ofile, delimiter=',')
    next(statsr)  # skip the header
    for row in statsr:
        if row[35] != season:
            season = row[35]
            if week != '-1' : print ('...', week, "weeks")
            print("Processing Season", season ,end='')
            week_count=0
        if row[39] != week:
            week = row[39]
            week_count = week_count + 1
            teams_seen = []
       
        if (row[40] == row[41]):
            print("\nteam playings itself in season,week,id:", season,week,row[40],end='')
    
        if (row[40] in teams_seen) & (row[40] != '-1'):
            print("\nFound repeated team within week in season,week,id:", season,week,row[40],end='')
        teams_seen.append(row[40])
        if (row[41] in teams_seen) & (row[41] != '-1') :
            print("\nFound repeated team within week in season,week,id:", season,week,row[41],end='')
        teams_seen.append(row[41])
        
        if (int(row[17]) > 0) & (int(row[2]) < 0) & (int(row[4]) < 0):
            print("\nsupicious game without yardage in season,week,id:", season,week,row[40],end='')
        
        if (int(row[34]) > 0) & (int(row[19]) < 0) & (int(row[21]) < 0):
            print("\nsupicious game without yardage in season,week,id:", season,week,row[41],end='')
        
        if int(row[17]) < 0:
            print("\nNegative Score?? season,week,id:", season,week,row[40],end='')
        
        if int(row[34]) < 0:
            print("\nNegative Score?? season,week,id:", season,week,row[41],end='')
        
        if int(row[17]) > 100:
            print("\nsuspiciously high score?? season,week,id:", season,week,row[40],end='')
        
        if int(row[34]) > 100:
            print("\nsuspiciously high score?? season,week,id:", season,week,row[41],end='')
           
        if int(row[6]) > int(row[5]):
            print("\nsupicious game more completions than attempts in season,week,id:", season,week,row[40],end='')
        
        if int(row[23]) > int(row[22]):
            print("\nsupicious game more completions than attempts in season,week,id:", season,week,row[41],end='')
            
            
    if week != '-1' : print ("...", week, "weeks")        
```
