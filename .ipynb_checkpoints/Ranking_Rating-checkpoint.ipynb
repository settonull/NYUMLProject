```python
import numpy as np
import pandas as pd
from datetime import *
```

```python
scores_data = pd.read_csv('data/NCAAAllScores2013.csv')
scores_data['Start'] = pd.to_datetime(scores_data['Start'])
scores_data['Date'] = scores_data['Start'].dt.date
scores_data.drop(scores_data.loc[scores_data['Date'] > date(2014, 1, 7)].index, axis=0, inplace=True)

sched_data = pd.read_csv('data/schedule2013.csv')
sched_data.rename({'Wk':'Week'}, axis=1, inplace=True)
sched_data['Date'] = pd.to_datetime(sched_data['Date']).dt.date
```

```python
def fix_dates(d, wrong=wrong_dates):
    if d in wrong_dates:
        return d - timedelta(days=1)
    else:
        return d
```

```python
wrong_dates = np.setdiff1d(scores_data['Date'].drop_duplicates().tolist(),sched_data['Date'].drop_duplicates().tolist())
scores_data['Date'] = scores_data['Date'].apply(lambda x: fix_dates(x))
```

```python
scores = scores_data[['Date','Away','Home','HomeFinal','VisFinal']]
sched = sched_data[['Week','Date']].drop_duplicates()
```

```python
scores_clean = pd.merge(scores, sched[['Date','Week']], on='Date',how='left')
scores_clean
```
